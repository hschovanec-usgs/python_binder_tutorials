## Basics
Numpy is an open source library that provides:
- Custom datatypes
- I/O with NumPy
- N-dimensional arrays
- Indexing
- Broadcasting
- Byte-swapping
- Structured arrays

We can begin by looking at the features through an example.

```python
import numpy as np
import matplotlib.pyplot as plt
```

There is a file called `sample.txt` located in the same folder as this notebook. This text file has comma delimited synthetic waveform data. There are four columns:
1. Time
2. North Component
3. East Component
4. Vertical Component 

We can begin looking at numpy I/O functionality by reading in this data. There are a two primary functions to read text data:
- loadtxt
- genfromtxt
loadtxt gives basic text reading ability, while genfromtxt expands upon this functionality to deal with incomplete datasets. It can fill missing data or replace data.

```python
waveforms = np.genfromtxt('sample.txt', delimiter=',')
```

```python
with np.printoptions(precision=3):
    print(waveforms)
```

If you look at the file you can see that there is no displacement data until line 2550. We can skip lines so that we are not holding all of the data in memory.

```python
waveforms = np.genfromtxt('sample.txt', delimiter=',', skip_header=2548)
with np.printoptions(precision=3):
    print(waveforms)
```

We can use slicing for numpy arrays, the same way we use it for lists, to split up the data.

```python
time = waveforms[:,0]
n = waveforms[:,1]
e = waveforms[:,2]
z = waveforms[:,3]
```

```python
plt.figure(figsize=(12, 15))
plt.subplot(3, 1, 1)
plt.plot(time, n)
plt.xlabel('Time (s)', fontsize=24)
plt.ylabel('Displacement (m)', fontsize=24)
plt.title('N', fontsize=30)

plt.subplot(3, 1, 2)
plt.plot(time, e)
plt.xlabel('Time (s)', fontsize=24)
plt.ylabel('Displacement (m)', fontsize=24)
plt.title('E', fontsize=30)


plt.subplot(3, 1, 3)
plt.plot(time, z)
plt.xlabel('Time (s)', fontsize=24)
plt.ylabel('Displacement (m)', fontsize=24)
plt.title('Z', fontsize=30)

plt.tight_layout()
plt.show()
```

```python
base = np.array([16, 5, 9, 4, 2, 11, 7, 14])

## Example 1
a = base[2:7]
print("Example 1:", a)

## Example 2
b = base[::2]
print("Example 2:", b)

## Example 3
c = base[::-1]
print("Example 3:", c)
```
